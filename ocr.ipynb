# Optical Character Recognition (ORC) 

```python
!pip install easyocr
import easyocr
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
Img = 'pic7.jpg'
reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(Img)
print(result)
```

# Original Image

```python
i = cv2.imread(Img)
plt.imshow(i)
plt.show()
```

Image showing Text

```python
coord = result[0][0]   # Co-ordinates
text = result[0][1]    # Text in the Image

# Taking the bounding values for creating a rectangle around the text.

x_min, y_min = [min(id) for id in zip(*coord)]  
x_max, y_max = [max(id) for id in zip(*coord)]    
image = cv2.imread(Img)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.rectangle(image, (x_min, y_min), (x_max,y_max), (0,255,0), 2)

# Showing  the text in the image above the rectangle.
image = cv2.putText(image, text,(x_min+30, y_min), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

plt.imshow(image)
plt.show()
```
